```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np


labels = ['Ensino Técnico', 'Ensino Médio', 'Ensino Superior']
sim = [3, 15, 1] 
nao = [5, 5, 0]

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, sim, width, label='Pratica esporte')
rects2 = ax.bar(x + width/2, nao, width, label='Não pratica esporte')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Quantidade')
ax.set_title('Mostra a distribuição de escolaridade e quem pratica esporte')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(2, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)

fig.tight_layout()

plt.show()
```
